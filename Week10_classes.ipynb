## Methods
You have already come across one special method, the init method. 
However, the power of classes will start to become more obvious to you as you begin writing your own custom methods. 
There are three types of methods you will explore in the following sections:

Instance methods; 
Static methods; 
Class methods

* **Instance methods**: the most common type of method you will need to use. They always take **self** as the first positional argument. 
* -- The __init__ method discussed in the previous section is an example of an instance method.
* **Static methods**: similar to instance methods, except that they do not implicitly pass the positional self argument.
* -- Static methods are defined using the @staticmethod decorator. (Decorators allow us to alter the behavior of functions and classes.)
* **Class methods**: Class methods are like instance methods, except that instead of the instance of an object being passed as the first positional self argument, the class itself is passed as the first argument. As with static methods, you use a decorator to designate a class method.  

```python
## Instance method: 
## self represents the instance (that is, the object) within the method. This is always the first positional argument of an instance method, 
## and Python passes it to the function without you needing to do anything. 

import math
class Circle:
    is_shape = True
## radius = 90
    def __init__(self, radius, color='red'):
        self.radius = radius
        self.color = color
    
    def area(self):
        return math.pi * self.radius ** 2

print(Circle(30, 'blue').radius)
print(Circle(30, 'blue').area())
 
```

```python
## Adding arguments to instance methods
####  Often, you need to specify other inputs to compute your methods. 
class Pet:
    def __init__(self, height):
        self.height = height
    
    is_human = False
    owner = 'Michael Smith'
    
    def is_tall(self, tall_if_at_least=30): # a new parameter 'tall_if_at_least is added
        return self.height >= tall_if_at_least

print(Pet(50).is_tall())
print(Pet(50).is_tall(tall_if_at_least=60))
```

```python
### The __str__ method
# Like __init__, the __str__ method is another special instance method that you need to know about.
# This is the method that is called whenever an object is rendered as a string.
class Pet:
    def __init__(self, height, name):
        self.height = height
        self.name = name
    
    is_human = False
    owner = 'Shawn Ha'
    
    def __str__(self):
        return f"{self.name} (height: {self.height} cm)"
print(type(Pet(40,'Chad')))
Pet(40,'Chad').owner
print(Pet(40,'Chad')) 
```

```python
###   Static methods
## Static methods are similar to instance methods, except that they do not implicitly pass the positional self argument.
##  Static methods are defined using the @staticmethod decorator. Decorators allow us to alter the behavior of functions and classes.
## Technically, this is actually passing the owned_by_smith_family function to a higher-order function that alters its behavior. 
class Pet:
    def __init__(self, height):
        self.height = height
    
    is_human = False
    owner = 'Michael Smith'
    
    @staticmethod  
    def owned_by_smith_family():
        return 'Smith' in Pet.owner
 
class Pet2:
    def __init__(self, height):
        self.height = height
    
    is_human = False
    owner = 'Michael Smith'
    
    ## @staticmethod
    def owned_by_smith_family( y ):
        return  y in Pet2.owner

##  pet vs pet2
nibbles = Pet(100)
nibbles.owned_by_smith_family()
nibbles.owner
nibbles.height
```

```python
nibbles2 = Pet2(100)

Pet2.owned_by_smith_family('Mitchell')
## nibbles2.owned_by_smith_family()   ##not under the instance nibbles2
nibbles2.owner
nibbles2.height
```

```python
## Class methods:   @classmethod
## Note that the first positional argument of this method is 'cls', not 'self'. You can call this method on the class itself:

class Australian:
    is_human = True
    enjoys_sport = True
    
    @classmethod
    def is_sporty_human(cls):
        return cls.is_human and cls.enjoys_sport
Australian.is_sporty_human()
aussie = Australian()
aussie.is_sporty_human()
```

```python
## size in kmsq - Country
class Country:
    def __init__(self, name='Unspecified', population=None, size_kmsq=None):
        self.name = name
        self.population = population
        self.size_kmsq = size_kmsq
## Suppose you want to avoid a situation where you create a country where people can specify the size in square miles rather than square kilometers.
## You could use a class method that takes the square mile input from the user and converts it into square kilometers, before initializing an instance of the class:
    @classmethod
    def create_with_msq(cls, name, population, size_msq):
        size_kmsq = size_msq / 0.621371 ** 2
        return cls(name, population, size_kmsq)
 
mexico = Country.create_with_msq('Mexico', 150e6, 760000)
mexico.size_kmsq
mexico = Country('Mexico', 150e6, 760000)
mexico.size_kmsq
```

## Properties
* Properties are designed with encapulation in mind (encapsulation is keeping tasks 'contained' in relevant namespaces)
 

```python
## interactive attributes
## What is wrong?
class Temperature():
    def __init__(self, celsius):
        self.celsius    = celsius
        self.fahrenheit = self.celsius * 9 / 5 + 32
        
temp = Temperature(100)
temp.fahrenheit      
temp.celsius
temp.celsius = 0
temp.fahrenheit
```

```python
## update the code using instance method
class Temperature():
    def __init__(self, celsius):
        self.celsius = celsius
    def fahrenheit(self):
        return self.celsius * 9 / 5 + 32

temp = Temperature(100)
temp.fahrenheit()
temp.celsius = 0
temp.fahrenheit()
```

This essentially solves the dynamic issue, but created a problem:

The obvious problem of additional syntax
A less obvious problem of potentially breaking your 'previous' or 'future' code
The 'previous' is an issue as you might have written your program thinking this would be an attribute and hence all uses would need to be updated to methods

The 'future' issue is the opposite, if you reorganize your class so that it needn't be a method anymore, you would have to change the references everywhere in your code

```python
## Both of these unsavory repercussions are addressed via properties, the core behavior of which is accessed via a decorator @property
class Temperature():
    def __init__(self, celsius):
        self.celsius = celsius
    @property  ## getter : When you access the attribute, this method will be called automatically. 
    def fahrenheit(self):
        return self.celsius * 9 / 5 + 32
    @fahrenheit.setter  ## setter: When you set the attribute, this method will be called automatically. 
    def fahrenheit(self,value):
        self.celsius = (value - 32) * 5 / 9
```

```python
temp = Temperature(100)
temp.fahrenheit
temp.fahrenheit = 32
temp.celsius
```
